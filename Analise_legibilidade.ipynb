```python
import pandas as pd
import os
import sys
from tqdm import tqdm
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
```

```python
pastes = os.listdir(".\\Dados limpos Juarez")

lista_dic = pd.read_csv("Palavras_dicionario.csv", header = None)
lista_dic = lista_dic[0].tolist()
```

```python
array = []
files_error = []
files_total = []
for paste in tqdm(pastes):
    files = os.listdir(f".\\Dados limpos Juarez\\{paste}")

    for i in files:
        a_file = open(f".\\Dados limpos Juarez\\{paste}\\{i}")

        list_of_lists = []
        flat_list = []

        for line in a_file:
            stripped_line = line.strip()
            line_list = stripped_line.split()
            list_of_lists.append(line_list)

            flat_list = [item for sublist in list_of_lists for item in sublist]


        
        words_ok = list(set(flat_list).intersection(lista_dic))
        words_total_qtd = len(flat_list)
        words_ok_qtd = len(words_ok)
        if len(flat_list) > 0:
            percent_words_ok = words_ok_qtd/len(flat_list) * 100
        else:
            percent_words_ok = 0

        array.append([i,words_ok,words_total_qtd, words_ok_qtd, percent_words_ok ])
        files_total.append(paste+'-'+i)
```

```python
df = pd.DataFrame(array)
df.columns = ['Arquivo', 'Palavras', 'Quantidade total de palavras','Quantidade de palavras legíveis', 'Porcentagem']
```

```python
df.to_csv('Juarez_words_full_qtd_all.csv', index = False)
```

```python
plt.figure(figsize=(10,6))

plt.hist(df['Porcentagem'], bins = 20)
plt.xlabel('Porcentagem')
plt.ylabel('Quantidade')
plt.title('Representação das porcentagens de palavras legíveis')
plt.show()
```
